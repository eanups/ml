```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, '/Users/anup.sethuram/DEV/LM/EanS/ml30/ml30/ml-ng/utils')
from multi_linear import zscore_normalize_features, run_gradient_descent_feng
```

```python
# create target data
x = np.arange(0, 20, 1)
y = x**2

# engineer features .
X = np.c_[x, x**2, x**3]
```

```python
X.shape
```

```python
model_w,model_b = run_gradient_descent_feng(X, y, iterations=10000, alpha=1e-7)

plt.scatter(x, y, marker='x', c='r', label="Actual Value"); plt.title("x, x**2, x**3 features")
plt.plot(x, X@model_w + model_b, label="Predicted Value"); plt.xlabel("x"); plt.ylabel("y"); plt.legend(); plt.show()
```
